```python
# import the necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Data/Antwoorden (1).xlsx') 
```

```python
df.head() 
```

```python
df.columns
```

```python
# lets make a word cloud of the column 'What's your field of study ?' 
from wordcloud import WordCloud

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = None, 
                min_font_size = 10).generate(' '.join(df["What's your field of study ?"]))
```

```python
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud)
```

```python
# lets analyze the column 'Familiarity with Robotics/AI\n(Scale 1-5)' 
df['Familiarity with Robotics/AI\n(Scale 1-5)'].value_counts().plot(kind='bar') 
```

```python
# Boxplot of the column 'Familiarity with Robotics/AI\n(Scale 1-5)'
sns.boxplot(df['Familiarity with Robotics/AI\n(Scale 1-5)']) 
#x-axis label
plt.xlabel('Familiarity with Robotics/AI')
plt.show()
```
